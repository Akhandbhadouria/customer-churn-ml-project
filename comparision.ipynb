```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    "col": ["Accuracy", "Precision", "Recall", "F1"],
    "knn": [0.847, 0.714, 0.368, 0.486],
    "tree": [0.852, 0.762, 0.358, 0.487]
}

df = pd.DataFrame(data)
x = np.arange(len(df["col"]))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, df["knn"], width, label="knn")
plt.bar(x + width/2, df["tree"], width, label="tree")

plt.xticks(x, df["col"])
plt.ylabel("Scolre")
plt.title("Classification Model Performance colmparison (Grouped Bar Chart)")
plt.legend()

plt.show()
```
